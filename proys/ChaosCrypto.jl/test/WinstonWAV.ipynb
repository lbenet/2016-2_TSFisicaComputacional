```julia
using Winston
using WAV
```

```julia
snd, sampFreq = wavread("440_sine.wav");
```

```julia
size(snd)
```

```julia
5060 / sampFreq
```

```julia
s1 = snd[:,1];
```

```julia
timeArray = (0:(5060-1)) / sampFreq
timeArray = timeArray * 1000 #scale to milliseconds
```

```julia
plot(timeArray, s1, kind="line", col="black", xlabel="Time (ms)", ylabel="Amplitude")
```
