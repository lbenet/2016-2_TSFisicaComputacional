```julia
#Para la varianza y el promedio

function medσ(R,N)
    x_prom= sum(R)/N #elpromedio
    R_1= [i-x_prom for i in R] #la resta del promedio con cada uno de los datos
    R_2=[i^2 for i in R_1] #la elevación al cuadrado
    σ_v=sum(R_2)/(N-1)  #la suma final y división entre N-1 para obtener la varianza    
    
    println(x_prom) ; println(σ_v)
end
```

```julia
#Así:
medσ(R,N)
```

```julia
mean(R) 
```

Coinciden! :D

```julia
function online_variance(R)
    n = 0
    mean = 0.0
    M2 = 0.0
     
    for x in R
        n += 1
        delta = x - mean
        mean += delta/n
        M2 += delta*(x - mean)
    end
    
    if n < 2
        return NaN
    else
        return  M2 / (n - 1)
        
        end 
end

```

```julia
online_variance(R)
```

```julia
@time(medσ(R,N))
```

```julia
@time(online_variance)
```

```julia
@time(var(R))
```

```julia
@time(mean(R))
```
