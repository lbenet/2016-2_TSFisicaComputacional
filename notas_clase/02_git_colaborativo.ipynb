# Colaborando con `git`

```
> git branch nombre_rama
> git branch -v
> git branch -d nombre_rama

> git branch lbenet/clase2
> git checkout lbenet/clase2

> # Las dos últimas instrucciones son equivalentes a:
> git checkout -b lbenet/clase2

> git checkout master
> git merge lbenet/clase2
```

Otra instrucción útil es 
```
> git reset HEAD <file>
```
y lo que hace es que regresa al último commit, o sea, "resetea" los cambios al último commit.
